#### Class weights for heavily imbalanced data set

wj=n_samples / (n_classes * n_samplesj)

Here,

wj is the weight for each class(j signifies the class)
n_samples is the total number of samples or rows in the dataset
n_classes is the total number of unique classes in the target
n_samplesj is the total number of rows of the respective class

```python
## Defining Class weights:
dir_cw_0 = len(y1) / (3*action_df[action_df['0']==-1].shape[0])
dir_cw_1 = len(y1) / (3*action_df[action_df['0']==0].shape[0])
dir_cw_2 = len(y1) / (3*action_df[action_df['0']==1].shape[0])

acc_cw_0 = len(y2) / (3*action_df[action_df['1']==0].shape[0])
acc_cw_1 = len(y2) / (3*action_df[action_df['1']==1].shape[0])

dec_cw_0 = len(y3) / (3*action_df[action_df['2']==0].shape[0])
dec_cw_1 = len(y3) / (3*action_df[action_df['2']==0.8].shape[0])
```

```python
DATADIR = os.getcwd()
PICSPATH = "cnn_model_data/pics/"
```

```python
h = 350
w = 600
rh = int(h/2)
rw = int(w/2)
training_data_1 = []
training_data_2 = []
training_data_3 = []
error_files = []
exp = []

def create_training_data():
    
    path = os.path.join(DATADIR, 'cnn_model_data/pics')
    
    for img in tqdm(os.listdir(path)):
        val = img.split('_')[0]
        try:
            img_array = cv2.imread(os.path.join(path, img), cv2.IMREAD_GRAYSCALE)
            crp_array = img_array[0:h, 0:w]
            new_array = cv2.resize(crp_array, ((rw,rh)))
            df_val = y_df.iloc[int(val)]
            training_data_1.append([new_array, 
                                  df_val[0], df_val[1], df_val[2]])
            training_data_2.append([new_array, df_val[3], df_val[4]])
            training_data_3.append([new_array, df_val[5], df_val[6]])
        except Exception as e:
            error_files.append(img)
            exp.append(e)
            pass

create_training_data()
print(len(training_data_1), len(training_data_2), len(training_data_3))
```

```python
random.shuffle(training_data_1)
random.shuffle(training_data_2)
random.shuffle(training_data_3)

X = []
y1 = []
y2 = []
y3 = []

for features, l, ndi, r in training_data_1:
    X.append(features)
    y1.append([l, ndi, r])
    
for features, na, ac in training_data_2:
    #X.append(features)
    y2.append([na, ac])
    
for features, ndc, dc in training_data_3:
    #X.append(features)
    y3.append([ndc, dc])
    
X = np.array(X)
y1 = np.array(y1)
y2 = np.array(y2)
y3 = np.array(y3)

X = X/255.0
```

```python
import tensorflow.contrib.keras as keras
import keras
from keras.models import Model
import tensorflow as tf
from tensorflow.keras.models import Sequential
from tensorflow.keras import layers
from tensorflow.keras.layers import Dense, Dropout, Activation, Flatten
from tensorflow.keras.layers import Conv2D, MaxPooling2D
from tensorflow.keras.callbacks import TensorBoard
from sklearn.model_selection import train_test_split
from sklearn.model_selection import RepeatedKFold
from keras.layers import Input
from numpy import mean
from numpy import std
from sklearn.utils import class_weight
from keras.callbacks import ModelCheckpoint
```

```python
car_input = Input(shape=(175, 300,1), name='Car')

x = layers.Conv2D(32, (3,3), padding="same", name = "Conv2d_1")(car_input)
x = layers.Activation("relu")(x)
x = layers.BatchNormalization()(x)
x = layers.MaxPooling2D(pool_size=(2,2))(x)
x = layers.Dropout(0.25)(x)

x = layers.Flatten(name="Flatten_1")(x)
x = layers.Dense(128)(x)
```

```python
direction_pred = layers.Dense(3, name='direction')(x)
direction_pred = layers.Activation("softmax", name="Direction_Output")(direction_pred)

dir_model = Model(
    inputs=car_input, 
    outputs=direction_pred)
```

```python
acceleration_pred = layers.Dense(2, name='acceleration')(x)
acceleration_pred = layers.Activation("sigmoid", name="Acceleration_Output")(acceleration_pred)

acc_model = Model(
    inputs=car_input, 
    outputs=acceleration_pred)
```

```python
deceleration_pred = layers.Dense(2, name='deceleration')(x)
deceleration_pred = layers.Activation("sigmoid", name="Deceleration_Output")(deceleration_pred)

dec_model = Model(
    inputs=car_input, 
    outputs=deceleration_pred)
```

```python
keras.utils.plot_model(dir_model, "direction_model_cnn.png", show_shapes=True)
```

```python
keras.utils.plot_model(acc_model, "acceleration_model_cnn.png", show_shapes=True)
```

```python
keras.utils.plot_model(dec_model, "deceleration_model_cnn.png", show_shapes=True)
```

```python
dir_model.compile(
    optimizer='Adam',
    loss=keras.losses.CategoricalCrossentropy(), 
    metrics=['accuracy'])

acc_model.compile(
    optimizer='Adam',
    loss=keras.losses.BinaryCrossentropy(from_logits=True), 
    metrics=['accuracy'])

dec_model.compile(
    optimizer='Adam',
    loss=keras.losses.BinaryCrossentropy(from_logits=True), 
    metrics=['accuracy'])
```
